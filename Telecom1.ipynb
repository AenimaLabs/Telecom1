##Extracción y limpieza de datos

```python
import numpy as np
import pandas as pd
import json
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ruta = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
```

```python
datos_api = requests.get(ruta)
```

```python
datos = json.loads(datos_api.text)
```

```python
pd.DataFrame(datos)
```

```python
datos_normalizados = pd.json_normalize(datos)
# Mostrar las primeras filas para contexto
print("\n=== Primeras 5 filas del dataset ===")
display(datos_normalizados.head())
```

```python
# Mostrar información general del DataFrame
print("=== Información general del dataset ===")
datos_normalizados.info()
```

```python
# Verificar tipos de datos por columna
print("\n=== Tipos de datos de cada columna ===")
datos_normalizados.dtypes
```

```python
# Verifique si hay valores únicos en cada columna
for col in datos_normalizados.columns:
  print(f"valores únicos en la columna'{col}': {datos_normalizados[col].nunique()}")
  if datos_normalizados[col].nunique() < 50: # Imprima los valores solamente si no hay muchos
    print(datos_normalizados[col].unique())
    print('-' * 50)  # Línea divisoria para mejor visualización
```

```python
# Eliminando las columnas de Churn

datos_normalizados = datos_normalizados[datos_normalizados['Churn'].str.strip() != '']
print("Número de filas después de eliminar las vacías en 'Churn':", len(datos_normalizados))
```

###Diccionario de datos
* customerID: número de identificación único de cada cliente
* Churn: si el cliente dejó o no la empresa
* gender: género (masculino y femenino)
* SeniorCitizen: información sobre si un cliente tiene o no una edad igual o mayor a 65 años
* Partner: si el cliente tiene o no una pareja
* Dependents: si el cliente tiene o no dependientes
* tenure: meses de contrato del cliente
* PhoneService: suscripción al servicio telefónico
* MultipleLines: suscripción a más de una línea telefónica
* InternetService: suscripción a un proveedor de internet
* OnlineSecurity: suscripción adicional de seguridad en línea
* OnlineBackup: suscripción adicional de respaldo en línea
* DeviceProtection: suscripción adicional de protección del dispositivo
* TechSupport: suscripción adicional de soporte técnico, menor tiempo de espera
* StreamingTV: suscripción de televisión por cable
* StreamingMovies: suscripción de streaming de películas
* Contract: tipo de contrato
* PaperlessBilling: si el cliente prefiere recibir la factura en línea
* PaymentMethod: forma de pago
* Charges.Monthly: total de todos los servicios del cliente por mes
* Charges.Total: total gastado por el cliente

```python
for column in datos_normalizados.columns:
    print(f"Unique values for '{column}':")
    print(datos_normalizados[column].unique())
    print("-" * 30)
```

```python
# 1. Renombrar todas las columnas: reemplazar '.' por '_'
df_clean = datos_normalizados.copy()
df_clean.columns = df_clean.columns.str.replace('.', '_', regex=False)

# 2. Opcional: renombrar manualmente para mayor claridad
columnas_finales = {
    'customer_gender': 'gender',
    'customer_SeniorCitizen': 'senior_citizen',
    'customer_Partner': 'partner',
    'customer_Dependents': 'dependents',
    'customer_tenure': 'tenure',
    'phone_PhoneService': 'phone_service',
    'phone_MultipleLines': 'multiple_lines',
    'internet_InternetService': 'internet_service',
    'internet_OnlineBackup': 'online_backup',
    'internet_DeviceProtection': 'device_protection',
    'internet_TechSupport': 'tech_support',
    'internet_StreamingTV': 'streaming_tv',
    'internet_StreamingMovies': 'streaming_movies',
    'account_Contract': 'contract',
    'account_PaperlessBilling': 'paperless_billing',
    'account_PaymentMethod': 'payment_method',
    'account_Charges_Monthly': 'monthly_charges',
    'account_Charges_Total': 'total_charges',
    'customerID': 'customer_id',
    'Churn': 'churn'
}

df_clean = df_clean.rename(columns=columnas_finales)

# 3. Verificar
print(df_clean.columns.tolist())
```
